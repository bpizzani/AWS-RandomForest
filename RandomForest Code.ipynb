```python
import datetime 

def splitter(x):
    if len(x.split("-")) == 2:
        return x.split("-")[1].upper()
    else:
        return x.split("-")[0].upper()

def splitdomain(x):
    return x.split(".")[0]

def splitdomain2(x):
    if len(x.split(".")) >= 3:
        return x.split(".")[2]
    elif len(x.split(".")) == 2:
        return x.split(".")[1]
    else:
        return x.split(".")[0]

def etix_paper(x):
    if x in ['PDF','PDF - PREDELIVERY','EXTERNAL MOBILE TRANSFER','BARCODE','BARCODE - PREDELIVERY (NON-STH)','BARCODE - PREDELIVERY (STH)']:
        return "ETICKET"
    else:
        return "HARDTICKET"

def instant(x):
    if x in ['PDF - PREDELIVERY','BARCODE - PREDELIVERY (NON-STH)','BARCODE - PREDELIVERY (STH)','OTHER - PREDELIVERY','LMS - PREDELIVERY']:
        return "INSTANT"
    else:
        return "NO_INSTANT"
def resolution(res):
    if (res == "Buyer Fraud  Loss") or (res == "Chargeback Received - Fraud") or (res == "ATO-Buyer Loss") or (res == "ATO-Buyer Prevented") or (res == "Chargeback - Hostile Fraud Missed"):
        return "Fraud"
    elif (res == "Buyer Fraud Prevented") :
        return "Suspicious"
    else:
        return "No Fraud"
def month(date):
    if str(date) >= "2019-07-00 00:00:00" and str(date) < "2019-07-31 23:59:59":
        return "July19"
    elif str(date) >= "2019-02-00 00:00:00" and str(date) < "2019-02-31 23:59:59":
        return "February19"
    elif str(date) >= "2019-01-00 00:00:00" and str(date) < "2019-01-31 23:59:59":
        return "January19"
    elif str(date) >= "2019-03-00 00:00:00" and str(date) < "2019-03-31 23:59:59":
        return "March19"
    elif str(date) >= "2019-04-00 00:00:00" and str(date) < "2019-04-31 23:59:59":
        return "April19"
    elif str(date) >= "2019-05-00 00:00:00" and str(date) < "2019-05-31 23:59:59":
        return "May19"
    elif str(date) >= "2019-06-00 00:00:00" and str(date) < "2019-06-31 23:59:59":
        return "June19"
    elif str(date) >= "2019-08-00 00:00:00" and str(date) < "2019-08-31 23:59:59":
        return "August19"
    elif str(date) >= "2019-09-00 00:00:00" and str(date) < "2019-09-31 23:59:59":
        return "September19"
    elif str(date) >= "2019-10-00 00:00:00" and str(date) < "2019-10-31 23:59:59":
        return "October19"
    elif str(date) >= "2019-11-00 00:00:00" and str(date) < "2019-11-31 23:59:59":
        return "November19"
    elif str(date) >= "2019-12-00 00:00:00" and str(date) < "2019-12-31 23:59:59":
        return "December19"
    elif str(date) >= "2020-01-00 00:00:00" and str(date) < "2020-01-31 23:59:59":
        return "January20"
    elif str(date) >= "2020-02-00 00:00:00" and str(date) < "2020-02-31 23:59:59":
        return "February20"
    elif str(date) >= "2020-03-00 00:00:00" and str(date) < "2020-03-31 23:59:59":
        return "March20"
    elif str(date) >= "2020-04-00 00:00:00" and str(date) < "2020-04-31 23:59:59":
        return "April20"
    elif str(date) >= "2020-05-00 00:00:00" and str(date) < "2020-05-31 23:59:59":
        return "May20"
    elif str(date) >= "2020-06-00 00:00:00" and str(date) < "2020-06-31 23:59:59":
        return "June20"
    elif str(date) >= "2020-07-00 00:00:00" and str(date) < "2020-07-31 23:59:59":
        return "July20"
    elif str(date) >= "2020-08-00 00:00:00" and str(date) < "2020-08-31 23:59:59":
        return "August20"
def ResolutionY(x):
    if (x == "Chargeback Received - Fraud".upper()) | (x == "Buyer Fraud  Loss".upper()) |  (x == "Chargeback - Hostile Fraud Missed".upper()):
        return "Fraud"
    elif x == "Buyer Fraud Prevented".upper():
        return "Suspicious"
    
    else:
        return "NoFraud"
def datetimer(x):
    if len(x["Import_Time"]) > 17:
        day = datetime.datetime.strptime(x["Event_Date_Time_Until"],'%Y-%m-%d')  - datetime.datetime.strptime(x["Import_Time"],'%Y-%m-%d') 
        return day.days
    else:
        day = datetime.datetime.strptime(x["Event_Date_Time_Until"],'%Y-%m-%d')  - datetime.datetime.strptime(x["Import_Time"],'%Y-%m-%d') 
        return day.days  
def dataprocessor(train):
    train["converted_transaction_amount"].fillna(train["converted_transaction_amount"].mean(),inplace=True)
    train["Buyer_Payment_CC_Type"].replace(" ", np.nan, regex=True,inplace=True)
    train["Buyer_Payment_CC_Type"].fillna("Paypal",inplace=True)
    train = train[train["Buyer_Payment_CC_Type"] != "Paypal"]
#train["BIN_Country"].fillna("Paypal")
#train= train[((train["Buyer_Payment_CC_Type"] == "MasterCard") | (train["Buyer_Payment_CC_Type"] == "Visa") | (train["Buyer_Payment_CC_Type"] == "AmericanExpress")) & (train["Recommendation_Code"] =="Review")]
    train["resolution1_name"].fillna("Accepted",inplace=True)
    train = train[(train["resolution1_name"]== "Cancelled - non fraud") | (train["resolution1_name"]== "Chargeback Received - Fraud") | (train["resolution1_name"]== "Buyer Fraud  Loss")  | (train["resolution1_name"]== "Fraud Reviewed  Allowed") | (train["resolution1_name"]== "Buyer Fraud Prevented") | (train["resolution1_name"]== "Accepted")  | (train["resolution1_name"]== "Chargeback - Hostile Fraud Missed")]
#
#train= train[(train["Recommendation_Code"] =="Review")]
    train= train[(train["Score_Total_Score"] >3800)]
#train = train[train["Buyer_IP_Continent"] == "north america"]
#train = train[train["lang"] == "US"]
#train = train[(train["Month"]==1) | (train["Month"]==2)]
#train = train[train["Month"]==3]
#train = train[train["Score_Total_Score"]< 10000]
#train =train[train["Event_Genre"] == "Concerts"]
#train = train[train["Rules_Tripped_in_RTD"].str.contains("5237260000001486029")]
    train["Buyer_IP_Country"].replace(" ", np.nan, regex=True,inplace=True)
    train["Buyer_IP_Country"]= train["Buyer_IP_Country"].fillna("Empty")
    train["ThreatMetrix_Browser_Languag~1"]= train["ThreatMetrix_Browser_Languag~1"].fillna("Empty")

    train["ThreatMetrix_Os"].replace(" ", np.nan, regex=True,inplace=True)
    train["ThreatMetrix_Os"]= train["ThreatMetrix_Os"].fillna("Empty")

    train["Event_Genre"].replace(" ", np.nan, regex=True,inplace=True)
    train["Event_Genre"]= train["Event_Genre"].fillna("Other/Sport")
    train["Venue_Country"]= train["Venue_Country"].replace("ENGLAND","GB")


    train["Resolution"] = train["resolution1_name"].apply(resolution)
    ISO_IP = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/IP country ISO2.xlsx",index_col=0)
    Genres = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/Event Genres.xlsx",index_col=0)
    Domains = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/Generic emails.xlsx",index_col=0)
    import sqlite3
    con = sqlite3.connect('C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/database.sqlite')
    cur = con.cursor()

# The result of a "cursor.execute" can be iterated over by row
    artists = pd.DataFrame(cur.execute('SELECT * FROM artists'))
    artists.columns = ["ID","Artist"]
    artists.index = artists["ID"]
    artists.drop("ID",axis=1,inplace=True)
    artists
    genres = pd.DataFrame(cur.execute('SELECT * FROM genres;'))
    genres.index = genres[0]
    genres.drop(0,inplace=True,axis=1)
    genres

    Artists_Genres = pd.merge(artists, genres, left_index=True, right_index=True)
    Artists_Genres = Artists_Genres.apply(lambda x: x.astype(str).str.upper())
    Artists_Genres.index= Artists_Genres["Artist"]
    Artists_Genres.drop("Artist",axis=1,inplace=True)

    Artists_Genres2 = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/artists with genres.xlsx",index_col=0)

    Capitals = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/City Capitals.xlsx",index_col=0)
    train= train.apply(lambda x: x.astype(str).str.upper())

    train["Month"] = train["Import_Time"].apply(month)
    
    train["Shared_Neg_IP"] = np.nan
    for i in train[train["Rules_Tripped_in_RTD"].str.contains("5237260000000356298")].index:
        train["Shared_Neg_IP"][i] = "HOT_IP"
    train["Shared_Neg_IP"].fillna("CLEAN_IP",inplace=True)
    train["Shared_Neg_IP"].value_counts()


    train["Shared_Neg_CC"] = np.nan
    for i in train[train["Rules_Tripped_in_RTD"].str.contains("5237260000000356289")].index:
        train["Shared_Neg_CC"][i] = "HOT_CC"
    train["Shared_Neg_CC"].fillna("CLEAN_CC",inplace=True)
    train["Shared_Neg_CC"].value_counts()

    train["Prepaid_CC"] = np.nan
    for i in train[train["Rules_Tripped_in_RTD"].str.contains("5237260000001436975")].index:
        train["Prepaid_CC"][i] = "PREPAID"
    train["Prepaid_CC"].fillna("CREDIT/DEBIT",inplace=True)
    train["Prepaid_CC"].value_counts()



    train["UNIVERSITY_IP"] = np.nan
    for i in train[train["Buyer_IP_Carrier_Code"].str.contains("UNIVERSI")].index:
        train["UNIVERSITY_IP"][i] = "UNIVERSITY"
    for i in train[train["Buyer_IP_Carrier_Code"].str.contains("JISC SERVICES LIMITED")].index:
        train["UNIVERSITY_IP"][i] = "UNIVERSITY"
    train["UNIVERSITY_IP"].fillna("NON_UNIVERSITY",inplace=True)


    train["Ticket_Format"]=train["Fulfillment_Method"].apply(etix_paper)
    train["Instant_Download"]=train["Fulfillment_Method"].apply(instant)

    
    train["Event_name"] = train["Listing_Genre_Description"].apply(lambda x: x.split(" TICKETS",1)[0].upper())

    train["Clean Domain"] = train["Buyer_Billing_Email_Domain"].apply(splitdomain)
    train["Domain Country"] = train["Buyer_Billing_Email_Domain"].apply(splitdomain2)
    train["Domain Category"] = train["Clean Domain"].map(Domains.to_dict()['GENERIC']).fillna("NO_GENERIC")

    train["lang"] = train["ThreatMetrix_Browser_Languag~1"].apply(splitter)
    train["ISO_IP"] = train["Buyer_IP_Country"].map(ISO_IP.to_dict()["IP_ISO"])

    train["cc_event"] = train["BIN_Country"] == train["Venue_Country"]
    train["cc_lang"] =train["lang"]== train["BIN_Country"]
    train["cc_ip"] =train["BIN_Country"] ==train["ISO_IP"]
    train["ip_event"] =train["ISO_IP"] == train["Venue_Country"]
    train["ip_lang"] = train["ISO_IP"] == train["lang"]
    train["ip_cc"] = train["ISO_IP"] == train["BIN_Country"]
    train["event_lang"]=train["Venue_Country"] ==train["lang"]
    #train["cardholder_buyer"] = train["payment_cc_holder_name"] == train["buyer_contact_fullname"]
    train["shipEbill"] = train["Buyer_Billing_Zip"] ==train["Buyer_Shipping_Zip"]
    train["primaryEbill"] = train["Buyer_Primary_Last_Name"] ==train["Buyer_Billing_Last_Name"]
    train["primaryEbill2"] = train["Buyer_Primary_First_Name"] ==train["Buyer_Billing_First_Name"]
    train["BillcountryECC"]  =train["BIN_Country"] ==train["Buyer_Billing_Country"]

    train["BillcityEevent"]  =train["Buyer_Billing_City"] ==train["Venue_City"]
    train["BillStateEevent"]  = train["Buyer_Billing_State"] ==train["Venue_State"]

    #train["IPstateEevent"]  = train["Buyer_IP_State"] ==train["Venue_State"]
    #train["IPstateEBill"]  = train["Buyer_IP_State"] ==train["Buyer_Billing_State"]

    train["IPcityEBill"]  = train["ThreatMetrix_True_IP_City"]==train["Buyer_Billing_City"]
    train["IPcityEevent"]  = train["ThreatMetrix_True_IP_City"] ==train["Venue_City"]

    train["IPcityEBill2"]  = train["ThreatMetrix_Input_IP_City"]==train["Buyer_Billing_City"]
    train["IPcityEevent2"]  = train["ThreatMetrix_Input_IP_City"] ==train["Venue_City"]

    train["billZipEIPZip"] = train["Buyer_Billing_Zip"] ==train["Buyer_IP_Zip"]
    #train["billZipEIPcity"] = train["Buyer_Billing_City"] ==train["Buyer_IP_Zip_City"]
    train["TrueIP_SourceIP"] = train["ThreatMetrix_True_IP"] ==train["Buyer_Source_IP"]




    train["IP_is_Cap"] = train["ThreatMetrix_True_IP_City"].map(Capitals.to_dict()["is capital"])
    train["IP_is_Cap"].fillna("N",inplace=True)

    train["Venue_is_Cap"] = train["Venue_City"].map(Capitals.to_dict()["is capital"])
    train["Venue_is_Cap"].fillna("N",inplace=True)

    #train["MusicGenre"] = train["Event_name"].map(Artists_Genres.to_dict()[1]).fillna("Other/Sports")
    train["MusicGenre"] = train["Event_name"].map(Genres.to_dict()["Genre"]).fillna(train["Event_name"].map(Artists_Genres.to_dict()[1]).fillna(train["Event_name"].map(Artists_Genres2.to_dict()["genre1"]).fillna("OTher/Sports")))



    train["Domain_CC"] = train["Domain Country"] == train["BIN_Country"]
    train["Domain_lang"] = train["Domain Country"] == train["lang"]
    train["Domain_event"] = train["Domain Country"] == train["Venue_Country"]
    train["Domain_ip"] = train["Domain Country"] == train["ISO_IP"]


  
    train["Event_Date_Time_Until"].dropna(inplace=True)
    train["Import_Time"].dropna(inplace=True)

    train["Daysleft"] = train[["Import_Time","Event_Date_Time_Until"]].apply(datetimer,axis=1)
    
    train["ResolutionY"] = train["resolution1_name"] .apply(ResolutionY)
    ISP = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/ISP.xlsx",index_col=0)
    ITSP = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/ITSP.xlsx",index_col=0)
    train["ISP_Category"] = train["Buyer_IP_Carrier_Code"].map(ISP.to_dict()["cat"]).fillna(train["Buyer_IP_Carrier_Code"].map(ITSP.to_dict()["cat"]))

    train["Event_name2"] = train["Event_Description"].apply(lambda x: x.split(" VS ",1)[0].upper())
    Football = pd.read_excel("C:/Users/bpizzanimendoza/Desktop/Python Michene Learning/Football Teams.xlsx",index_col=0)
    train["Football_Host"] = train[train["Event_Genre"] == "OTHER/SPORT"]["Event_name"].map(Football.to_dict()["Country"])
    train["Football_Visitor"] = train[train["Event_Genre"] == "OTHER/SPORT"]["Event_name2"].map(Football.to_dict()["Country"])
    train["Host_Visit"] = train["Football_Visitor"]  ==  train["Football_Host"] 

    train["Visit_CC"]  = train["Football_Visitor"]  == train["BIN_Country"]
    train["Visit_IP"]  = train["Football_Visitor"] ==train["ISO_IP"]
    train["Visit_lang"]  = train["Football_Visitor"] ==train["lang"]

    train["Host_CC"]  = train["Football_Host"]  == train["BIN_Country"]
    train["Host_IP"]  = train["Football_Host"] ==train["ISO_IP"]
    train["Host_lang"]  = train["Football_Host"] ==train["lang"]
    U_venue_country = train["Venue_Country"].unique()
    U_venues = train["Venue_Name"].unique()
    U_CCbin= train["BIN_Country"].unique()
    u_domains = train["Buyer_Billing_Email_Domain"].unique()

    U_ISP =  train["Buyer_IP_Carrier_Code"].unique()
    U_Genre = train["MusicGenre"].unique()

    U_os = train["ThreatMetrix_Os"].unique()
    U_CC_type = train["Buyer_Payment_CC_Type"].unique()

    U_Genres2 = train["Event_Genre"].unique()

    U_FM = train["Fulfillment_Method"].unique()

    U_Ips = train["ThreatMetrix_True_IP"].unique()

    train["Connectione_wIRELESS"] = train["Buyer_IP_Connection_Type"].apply(lambda x: x == "MOBILE WIRELESS")
    train["IP_seen"] = train["ThreatMetrix_True_IP"].map(train["ThreatMetrix_True_IP"].value_counts().to_dict())



# different users per IP
    d = train.pivot_table(index=["ThreatMetrix_True_IP","Buyer_Billing_Email"],aggfunc="count")
    email_events0 = {}
    for email in train["ThreatMetrix_True_IP"].unique():
            email_events0.update({email:len(d.loc[email].index)})
    train["Count_Users_per_IP"] = train["ThreatMetrix_True_IP"].map(email_events0)     


#different events per user
    d = train.pivot_table(index=["Buyer_Billing_Email","Event_name"],aggfunc="count")
    email_events1 = {}
    for email in train["Buyer_Billing_Email"].unique():
            email_events1.update({email:len(d.loc[email].index)})
    train["Count_Events_per_Email"] = train["Buyer_Billing_Email"].map(email_events1)   

#different events genre per user
    d = train.pivot_table(index=["Buyer_Billing_Email","Event_Genre"],aggfunc="count")
    email_events2 = {}
    for email in train["Buyer_Billing_Email"].unique():
            email_events2.update({email:len(d.loc[email].index)})
    train["Count_Genres_per_Email"] = train["Buyer_Billing_Email"].map(email_events2)     

#differets CCs country per User
    d = train.pivot_table(index=["Buyer_Billing_Email","BIN_Country"],aggfunc="count")
    email_events3 = {}
    for email in train["Buyer_Billing_Email"].unique():
            email_events3.update({email:len(d.loc[email].index)})
    train["Count_BIN_per_Email"] = train["Buyer_Billing_Email"].map(email_events3)  


#differets CC type  per User
    d = train.pivot_table(index=["Buyer_Billing_Email","Buyer_Payment_CC_Type"],aggfunc="count")
    email_events4 = {}
    for email in train["Buyer_Billing_Email"].unique():
        email_events4.update({email:len(d.loc[email].index)})
    train["Count_cctype_per_Email"] = train["Buyer_Billing_Email"].map(email_events4)  


#differets CCs country per IP
    d = train.pivot_table(index=["ThreatMetrix_True_IP","BIN_Country"],aggfunc="count")
    email_events5 = {}
    for email in train["ThreatMetrix_True_IP"].unique():
            email_events5.update({email:len(d.loc[email].index)})
    train["Count_BIN_per_IP"] = train["ThreatMetrix_True_IP"].map(email_events5) 

#differets Events per IP
    d = train.pivot_table(index=["ThreatMetrix_True_IP","Event_name"],aggfunc="count")
    email_events6 = {}
    for email in train["ThreatMetrix_True_IP"].unique():
            email_events6.update({email:len(d.loc[email].index)})
    train["Count_Events_per_IP"] = train["ThreatMetrix_True_IP"].map(email_events6) 

    #Buyer_IP_Country"
    cat_feats =["Fulfillment_Method","ThreatMetrix_Os","Guest_Order_Indicator","Event_Genre","Buyer_Payment_AVS_Zip_Result","Buyer_Payment_CC_Type","StubHubcom_Buyer_Profile_Sou~1","shipEbill", "primaryEbill" , "BillcountryECC","Buyer_IP_Continent","event_lang","ip_cc","ip_lang","ip_event","cc_ip","cc_lang","cc_event","BillcityEevent","BillStateEevent","IPcityEBill","IPcityEevent","MusicGenre","IPcityEBill2","IPcityEevent2","billZipEIPZip","Domain Category","Domain_CC","Domain_lang","Domain_event","Domain_ip","Ticket_Format","Instant_Download","UNIVERSITY_IP","Shared_Neg_IP","ISP_Category","Shared_Neg_CC","TrueIP_SourceIP","Genre_seen_before_with_this_~1","Host_Visit","Visit_CC","Visit_IP","Visit_lang","Host_CC","Host_IP","Host_lang","Book_of_Business","Prepaid_CC","primaryEbill2","BIN_Country", "Venue_Country","Connectione_wIRELESS","IP_is_Cap","Venue_is_Cap"]
    new_df = pd.get_dummies(train,columns=cat_feats,drop_first=True)
    
    dummies = new_df.iloc[:,133:]
    non_dummies = new_df[["ResolutionY","Month","Daysleft","Score_Accertify_Index","Email_Age_Score",'Count_Times_Billing_CC_seen_~1',
 'Count_Times_Billing_Email_se~1',"Listing_Number_Tickets_Purch~1","Number_of_Orders_Purchased_1~1","Buyer_Join_to_Transaction_Pu~1","Model_Probability","Distance_Buyer_IP_Zip_:_Buye~1","IP_seen","Count_Events_per_Email","Count_Genres_per_Email","Count_BIN_per_Email","Count_BIN_per_IP","Count_Users_per_IP","Count_cctype_per_Email","Count_Events_per_IP"]]
    train = pd.concat([non_dummies,dummies],axis=1)
    #train = train.drop(["Month","ResolutionY"],axis=1)
    
    return train
```
